## Parte 3.3: Relaci칩n radio de Einstein con la masa de las galaxias

Como se indic칩 anteriormente, el radio de Einstein nos da la capacidad de medir la masa de la galaxia lente. Los par치metros con los que hemos jugado est치n relacionados con la geometr칤a entre las dos galaxias y nosotros y con el radio de Einstein en particular. Al optimizar los par치metros en las 칰ltimas dos actividades, calculaste el valor estimativo del radio de Einstein. Sabiendo esto y si tuvi칠ramos las diversas distancias involucradas, se podr칤a usar la siguiente f칩rmula para encontrar la masa de la galaxia lente.

\\

$$ \theta_1 = \sqrt{\frac{4GM}{c^2}\frac{D_{LS}}{D_LD_S}} $$

En la ecuaci칩n anterior:

>$\theta_1$ es el radio de Einstein   
>$c$ es la velocidad de la luz.  
>$G$ es la Constante Gravitacional.  
>$M$ es la masa de la galaxia Lente (Chris).  
>$D_L$ es la distancia a la galaxia Lente (Chris).  
>$D_S$ es la distancia a la galaxia Amplificada (Pat).  
>$D_{LS}$ es la distancia desde la galaxia Amplificada (Pat) a la galaxia Lente (Chris).

Las distancias a estas galaxias est치n determinadas por los respectivos espectros desplazados hacia el rojo de las galaxias.
